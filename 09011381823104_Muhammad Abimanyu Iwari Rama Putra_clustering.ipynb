<H1> CLUSTERING DENGAN METODE K MEANS DAN PCA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.datasets import load_diabetes
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA # Principal Component Analysis module
from sklearn.cluster import KMeans # KMeans clustering 
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
%matplotlib inline
```

```python
#loads breast cancer dataset into variable by name cancer. 
diabetes = load_diabetes() 
# creating dataframe 
di = pd.DataFrame(diabetes['data'], columns = diabetes['feature_names']) 
# checking head of dataframe 
di.head()
```

```python
X = diabetes.data[:, :2]
y = diabetes.target
print (X)
```

```python
print (y)
```

```python
plt.scatter(X[:,0], X[:,1], c=y, cmap='gist_rainbow')
plt.xlabel('age', fontsize=18)
plt.ylabel('bmi', fontsize=18)
```

```python
km = KMeans(n_clusters = 2, random_state=21)
km.fit(X)
```

```python
centers = km.cluster_centers_
print(centers)
```

```python
new_labels = km.labels_
print (new_labels)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16,10))
axes[0].scatter(X[:, 0], X[:, 1], c=y, cmap='gist_rainbow',edgecolor='k', s=150)
axes[1].scatter(X[:, 0], X[:, 1], c=new_labels, cmap='jet',edgecolor='k', s=150)
axes[0].set_xlabel('age', fontsize=18)
axes[0].set_ylabel('bmi', fontsize=18)
axes[1].set_xlabel('age', fontsize=18)
axes[1].set_ylabel('bmi', fontsize=18)
axes[0].tick_params(direction='in', length=10, width=5, colors='k', labelsize=20)
axes[1].tick_params(direction='in', length=10, width=5, colors='k', labelsize=20)
axes[0].set_title('Actual', fontsize=18)
axes[1].set_title('Predicted', fontsize=18)
```

```python
from scipy.spatial.distance import cdist 
distortions = [] 
inertias = [] 
mapping1 = {} 
mapping2 = {} 
K = range(1,10) 
  
for k in K: 
    #Building and fitting the model 
    kmeanModel = KMeans(n_clusters=k).fit(X) 
    kmeanModel.fit(X)     
      
    distortions.append(sum(np.min(cdist(X, kmeanModel.cluster_centers_, 'euclidean'),axis=1)) / X.shape[0]) 
    inertias.append(kmeanModel.inertia_)   
    mapping1[k] = sum(np.min(cdist(X, kmeanModel.cluster_centers_, 'euclidean'),axis=1)) / X.shape[0] 
    mapping2[k] = kmeanModel.inertia_ 
```

```python
for key,val in mapping1.items(): 
    print(str(key)+' : '+str(val)) 
```

<h4> K MEANS ELBOW METHOD
    

```python
plt.plot(K, distortions, 'bx-') 
plt.xlabel('Values of K') 
plt.ylabel('Distortion') 
plt.title('The Elbow Method using Distortion') 
plt.show() 
```

<h1>Principal Component Analysis (PCA)
    
<h4>Mengimplementasikan modul PCA sklearn.

```python
scalar = StandardScaler() 
  
# Standardizing 
scalar.fit(di) 
scaled_data = scalar.transform(di) 
  
# applying PCA
pca = PCA(n_components = 3) 
pca.fit(scaled_data) 
x_pca = pca.transform(scaled_data)
x_pca.shape
```

<h4> 2D RESULTS

```python
plt.figure(figsize =(10, 8)) 
  
plt.scatter(x_pca[:, 0], x_pca[:, 1], c = diabetes['target']) 

# labeling x and y axes 
plt.xlabel('First Principal Component') 
plt.ylabel('Second Principal Component')
```

<h4> 3D RESULTS

```python
ax = plt.axes(projection='3d')
ax.scatter(x_pca[:, 0], x_pca[:, 1], x_pca[:, 2], c=diabetes['target'], cmap='viridis', linewidth=1);
```
